## 언어-구글 번역기

### 1. 개요
*   구글 번역기가 첫 선을 보인 것은 2006년이다. 
*   현재 전 세계 많은 사람들이 구글 번역기를 사용하고 있고,      
매일 몇 천억 개의 단어가 구글 번역기를 통해 번역되고 있다.

### 2. 구글 번역기의 변화
*   과거 구글 번역기는 기계학습으로 언어들을 번역했었다.   
다시 말해, 단어 하나하나를 번역하고, 그 단어들을 조합해서 문장을 만들었었다. 그래서 문장이 어색하기도 하고 잘못 번역되기도 하였다. 게다가 한글과 영어처럼 문장구조가 많이 다를 때에는 엉뚱한 번역을 하기도 하였다.
*   그러나 구글이, 인공신경망 ‘딥러닝 기술’을 번역기에 적용하면서부터
현재 구글 번역기는 예전보다 훨씬 자연스러워졌고 정확도를 높이게 되었다.   
업그레이드 된 구글 번역기는, 지금까지 번역된 글과 원문들을 찾아서, 단어와 문장 사이에 상관관계를 파악해 인공지능 스스로 번역 규칙을 찾고 번역을 한다. 다시 말해, 단어를 조합하는 것이 아니라 
문맥을 파악하고 문장을 번역하는 것이다.   
여기서 우리는 전체 문장들 사이의 상관성을 찾아 결과를 출력하는 것으로 보아 언어 번역에 ‘딥러닝 방식’이 사용되었음을 알 수 있다.

### 3. 구글 번역기의 인공지능 기술(딥러닝 방식)
*   딥러닝 방식이란 대량의 데이터와 기술을 활용하여,
 신경망의 계층을 다양한 방식으로 연속되게 구조화한 방식을 말한다.
 이때 신경망들은 각기 서로 결합하기도 하고, 교차하기도 하고, 반복하기도 하면서, 다양한 방식으로 구조화된다.
*   기본적으로 신경망은 인간의 두뇌를 모방한 것으로,
 기계가 입력 값에 대해 여러 단계의 신경망을 거쳐, 스스로 생각하고 판단하는 시스템이다.
*   머신러닝과 딥러닝 방식을 간단히 비교해보자면,   
 머신러닝은 먼저 데이터로 학습을 시키고, 그 학습을 바탕으로 데이터를 판단한다면, 딥러닝은 데이터만 제공을 해줘도, 스스로 규칙을 찾아서 학습해나간다는 차이점이 있다.   
 즉, 인공신경망 이론을 기본으로 사람이 모든 판단 기준을 정해주지 않아도, 컴퓨터가 스스로 인지, 추론, 판단할 수 있게 된다는 의미이다.
*   그래서 딥러닝 방식은 음성, 이미지 인식과 사진 분석 등에 광범위에게 활용이 된다.   
 특히 언어 번역 부분에서 두드러지는 특징을 나타내고 있다.

### 4. 구글 번역기의 발전
*   업그레이드 된 새 번역 시스템은, 구글 번역기가 출시된 이래,   
 지난 약 15년여 간 이룩한 발전에 맞먹는 성능개선을 이뤄냈다.
*   업그레이드 된 새 번역 시스템은 한층 향상된 문법과 구문을 제공할 수 있게 되었는데,
 개선 전과 비교해서 오류가 55~85% 가량 줄었다고 한다.
 그리고 구글 앱에서 지원하는 103개 언어 중 8개 언어에 업그레이드 버전이 적용되었는데, 영어, 프랑스어, 독일어, 스페인어, 중국어, 일본어, 터키어 그리고 한국어가 대상이다.
*   또 구글은 법륜 스님 강연을 영어로, 수학능력시험 영어문제 지문을 한글로 번역하는 모습 등을 시연했었다. 게다가 스마트폰 카메라를 맞추면, 모르는 언어로 된 간판 문구를 번역해주는 모습도 보여줬다. 그리고 인공신경망 인공지능기술 적용범위를 확대해 사진검색 및 스캔 서비스 기능까지 내놓았다.

### 5. 구글 번역기의 미래 전망
*   인공지능은 사람이랑 꽤 닮은 부분이 있다.   
 아이가 부모 행동을 보고 똑같이 따라하는 것처럼 인공지능도 사람한테 배운 대로 행동을 한다. 그래서 인공지능의 학습된 프로그램은 인간보다 훨씬 빠르게 해결해 나가기 때문에, 인간보다 뛰어난 일을 해낼 수 있게 된다.   
그러나 잘못된 말이나 편견이 담긴 말을 가르치게 되면 그런 말들을 똑같이 번역하기 때문에 조심해야 한다.
 우리는 인공지능의 도움을 받기도 하지만 우리가 인공지능에게 가르쳐야 하는 위치이기도 하기 때문이다.    
 그렇기 때문에 우리는 인공지능의 좋은 도움을 받기 위해서는 어떤 말들을 가르쳐야 할까 고민해봐야 한다.   



## 음성-인공지능 스피커(sk텔레콤 누구 스피커)

### 1. 개요
*   과거 음성인식의 역할은 소리를 문자로 바꾸어 주는 것에 한정되어 있었으나, 현재는 사람과 기계간의 대화를 가능하게 하는 자연어 처리 기술까지 발전해왔다.

### 2. 인공지능 스피커의 인공지능 기술(딥러닝 방식)
*   음성인식 기술이 오늘날 우리 일상생활에 가까이 다가오고 있는 배경에는, 하드웨어의 비약적 발전이 자리하고 있다.
 다시 말해 최근 몇 년 동안 딥러닝으로 말미암아 혁신적인 성능을 보이고 있는 음성인식 기술이 보편화 되고 있는 것은, 하드웨어의 눈부신 발전덕분이다.   
 게다가 최근 음성인식에 사용되는 주요한 딥러닝 기술 패러다임들은, 
 고성능 GPU와 같은 하드웨어의 지원에 편승하여, 비로소 꽃을 피우게 되었다.
*   인공지능 스피커는 딥러닝 기술을 사용하는데, 다양한 영역의 데이터가 누적됨에 따라, 스스로 학습해서 지식을 구축해 나아간다. 즉 인간의 뇌처럼 스스로 학습하고 지식을 축적해 나가는 형태로 역할을 한다는 뜻이다.
*   딥러닝 기술은 데이터만 주어지면 인공지능이 스스로 학습해서 지식을 쌓아가는 형태인 자율학습 방법론을 취하기 때문에, 사람의 개입은 최소화된다.  
 게다가, 하나의 지식이 생성되면 유사한 지식을 스스로 확장해 나가고,
 다양한 지식을 검색하고 분석하는 방식에서 나아가, 기존 지식에 기반해서 새로운 사실을 예측해 나간다. 예를 들어 음성 신호에서 특정 어휘를 인식한 다음, 단어가 문맥적으로 어떤 의미를 내포하고 있는지, 어떤 감정이 포함되어 있는지까지 파악하여 대화를 진행할 수 있다.

### 3. 인공지능 스피커 ‘sk 텔레콤 누구’ 특징
*   인공지능 스피커 sk 텔레콤 누구만의 차별요소는, 클라우드 환경에서 빅데이터를 이용하는 것이다. 즉, 딥러닝 기술을 기반으로 하기 때문에,
 데이터가 늘어날수록 데이터가 축적이 되면서 사용자의 음성 인식률이 높아지는 구조이다.
*   음악검색과 추천, 교통정보, 배달주문, 스마트홈 연동, 개인비서, t맵 연결, 날씨검색 등 매우 다양한 기능들이 지원되며 계속해서 화장되고 있다.

### 4. 인공지능 스피커의 미래 전망
*   음성인식 기술을 위해 딥러닝 알고리즘과 GPU와 같은 하드웨어 혁신은 계속해서 지속되어야 하며, 이를 통해 인류는 보다 나은 미래를 맞이할 수 있을 것이다.
*   아직 음성인식 기술이 완전하다고는 할 수 없지만, 음성인식 성공의 또 다른 측면에는 계속해서 포기하지 않는 연구자들의 끊임없는 노력이 있었다.
*   인공지능 기술은 계속해서 발전해 나갈 것이며, 그와 더불어서 주변의 많은 사물들도 더욱 스마트해질 것이다. 본격적인 사물인터넷 시대와 제로 시대가 도래하고 있다. 이에 대한 적절한 대응이 반드시 필요할 것이다.




## 이미지-인공지능 안면인식(Alexnet)
### 1. 개요
*   인공지능 안면인식은 먼저 디지털 이미지에서 얼굴을 검출한 뒤에,
 얼굴에서 눈, 코, 입, 점, 주름 등의 상대적 위치를 찾아낸 후,
 기존 정보와 매칭된 얼굴을 찾아내고 인식하는 방식이다.

### 2. 인공지능 안면인식이 사용되는 다양한 사례들
*   안면인식 기술은 일상생활에서 광범위하게 사용되고 있다.   
 우선 우리 일상에서는 생체인식 기술로 스마트폰의 잠금해제와 결제 시스템에서 주로 사용되고 있다. 또 마케팅 분야에서는 광고나 디스플레이를 보고 있는 대상의 성별과 연령대를 구분하여 적절한 광고를 보여주기도 한다. 예를 들어 아이와 함께 가족이 걸어가고 있다면 장난감이나 근처의 식당을 소개하고, 젊은 연인들이 걸어가고 있다면 커피숍이나 추천 데이트 코스를 보여준다.
*   또한 아직 완벽하지는 않지만 CCTV로 찍은 디지털 이미지에서 범죄자의 얼굴을 구별할 수도 있다. 게다가 갑자기 뛰는 사람이나 군중집회처럼 비일상적인 활동도 식별이 가능하기 때문에, 반사회적인 행위를 미연에 감지해서 추적이 가능하다.
*   마지막으로 최근에는 비행기에 탑승하는 승객들의 신원확인 절차를 안면인식 기능으로 진행했다고 한다.   
 이는 4차산업혁명을 대표하는 인공지능 안면인식 기술이 비행기 분야까지 확장된 사례임을 보여준다.

### 3. 인공지능 안면인식의 인공지능 기술(딥러닝)
*   딥러닝은 개인을 고유하게 식별하거나 확인할 수 있는 생체 인식 소프트웨어를 구축할 수 있는 힘을 제공한다.
*   딥러닝 방식은 이미지 결과를 이전에 축적된 데이터와 비교하고,   
 이미지에서 얼굴을 결정한 후에는, 최적의 이미지를 선택하기 위해 계속 반복한다. 즉 획득한 데이터를 데이터베이스에서 사용가능한 데이터와 비교하고, 매개변수가 일치하면 사람을 식별하는 방식이다.
*   머신러닝 알고리즘은 선형이지만, 딥러닝 알고리즘은 복잡성과 추상화가 증가하는 계층 구조이다. 모르는 것이 있다면 이전 계층에서 얻은 지식으로 계층을 구축하여 복잡한 추상화를 명확히 한다.   
 계층 구조의 각 알고리즘은 입력에 비선형 변환을 적용하고, 학습한 내용을 사용하여, 통계모델을 출력으로 만든다. 출력이 허용 가능한 수준의 정확도에 도달할 때까지 반복하여 계속한다.
*   딥러닝 방식은 매우 큰 얼굴의 데이터 세트를 활용하고, 풍부하고 간결한 얼굴 표현을 학습할 수 있기 때문에, 나중에 사람의 얼굴 인식 기능을 능가할 수도 있을 것이다.

### 4. 인공지능 안면인식의 미래 전망
*   안면인식은 우리 일상생활에서 생체보안 뿐만 아니라,  CCTV의 범죄자를 찾아내는 것과 같은 분야까지도 널리 적용되고 있다.   
 인공지능 안면인식 기술이 생체 보안을 넘어 감정 인식의 분야까지 확대된다면, 비즈니스 분야뿐만 아니라 일상생활에 큰 문제를 일으킬 수도 있을 것이다.
*   그러나 한 가지 확실한 것은 시간이 지날수록 안면인식 기술은 더욱 발전할 것이다.


## 자율주행-무인 자동차(자율주행 자동차 올리)
### 1. 개요
*   신산업 분야에서 가장 많이 주목받고 있는 분야 중 하나가, 바로 '자율주행 자동차'이다. 그 중 올리는, 로컬 모터스(Local Motors)에서 IBM의 인공지능을 활용해 제작한 자율주행 전기버스이다.
*   여기서 자율주행 자동차란 운전자가 직접 조항 및 변속장치를 조작하지 않아도, 설정된 목적지까지 스스로 이동하는 차량을 의미한다.

### 2. 무인 자동차의 기본원리와 3가지 주요기능
*   무인 자동차의 기본원리와 구조는 로봇과 매우 유사하다.    
 기본적으로 세 가지의 주요 장치가 있는데, 센서부, 제어부 그리고 동작장치로 이루어진다.
 먼저 센서부는 카메라 등을 이용하여 주변의 정보를 수집해서, 실시간으로 제어부에 전달한다. 그러면 제어부는 수집된 정보를 분석하고 판단하여, 자동차의 다음 동작을 결정하고 동작장치에 알린다. 마지막으로 동작 장치에 알리면, 그 지시에 따라 정확하게 차를 움직이게 된다.
*   이밖에도 자율주행 자동차를 구현하는 3가지 주요기능이 있는데,   
인지기능, 판단기능, 제어기능이다. ‘인지’ 기술을 통해 도로의 상황 및 현재 위치 그리고 지형·지물을 감지하고, 수집 된 정보를 ‘판단 및 분석’하여, 목적지까지 안전하게 도달할 수 있도록 ‘제어’하는 것이 바로 자율주행의 기본원리이다.

### 3. 무인 자동차의 인공지능 기능(딥러닝)
*   자율주행 자동차를 말할 때 빠지지 않고 등장하는 기술 중 하나가, 딥러닝이다.
*   딥러닝을 활용하고 특정 기능을 인공지능으로 구현하기 위해서는,   
 데이터를 반복적으로 입력하는 과정을 통해, 컴퓨터가 필요한 기능을 스스로 학습해가도록 해야 한다.
 즉 딥러닝 기반 인공지능 구현을 위해서 필요한 것은,
 기능에 대한 전문적인 지식보다는, 컴퓨터를 충분히 학습시킬 수 있는 데이터이다.
*   딥러닝은 자율주행 자동차를 구현하는 3가지 주요기능(인지기능, 판단기능, 제어기능)에서 사용되는데, 특히 카메라를 이용한 주행환경 ‘인지’분야에서 딥러닝은 가장 중요한 기술로 자리 잡았다.   
 카메라를 통해 입력된 이미지에 딥러닝을 적용하면, 자율주행 시스템에 필요한 환경정보들을 전부 검출하고 분류하는 기능을 한다. 또한 기존에는 인식이 불가능하다고 여겨졌던 영역까지도 점차 개선되고 있다.
*   그리고 딥러닝은 차량의 궤적을 생성하는 ‘판단’기능에서도 사용된다.
 행동 방식을 논리적으로 파악하기 어려울 때 딥러닝은 좋은 솔루션이 되는데, 다양한 운전방식과 관련 센서정보를 데이터로 입력한 후, 이를 딥러닝 알고리즘으로 학습한다면, 정확한 수학 모델 없이 데이터만으로도, 다른 운전자의 운행을 예측할 수 있는 인공지능 구현이 가능하다.
*   마지막으로 자량 ‘제어’분야에서도 딥러닝이 사용된다. 고도의 안전성이 요구되는 차량 제어기능을 구현하려면, 검증된 기존 기술을 사용하는 게 일반적이나, 딥러닝 기술을 활용해 개개인의 운전방식을 데이터화하면, 인간 감성을 고려한 차량 제어가 가능해진다.   
 다시 말해 단순 자율주행 기능뿐 아니라, 탑승자의 상태, 음성 등을 인식해서, 개별 탑승자에게 맞춤형 편의기술을 제공하는 서비스에도 딥러닝이 적용된다는 것이다.
*   딥러닝에서 개발한 인공지능들은 대체로 학습을 통해서 점점 더 발전되고 사고능력이 좋아지는 경향이 있으며, 이와 같은 인공지능기술이 자율주행 자동차의 핵심적인 부분으로 자리 잡을 가능성이 높아지고 있다.

### 4. 무인 자동차의 미래 전망
*   자동차가 움직이려면 연료가 있어야 하는 것처럼
 자율주행 자동차가 움직이려면 딥러닝 기술 구현에 활용될 데이터 구축이 필수적이다. 자동차는 도로를 주행하며 다양한 센서를 동원해서 빅데이터를 형성할 테고, 이 데이터는 클라우드 서비스로 공유된 후 자율주행 자동차의 딥러닝 알고리즘 학습에 쓰일 것이다.   
 빅데이터를 통합 딥러닝 학습은 한층 발전된 자율주행을 가능케 할 것이며,
 자율 주행으로 생성된 빅데이터는 다시 공유, 학습될 것이다. 이와 같은 데이터와 학습의 선순환 구조가 자율주행 자동차를 더 안전하게 만들길 기대한다.
*   딥러닝을 통한 인공지능 기술의 발전은, 앞으로 자율주행 자동차의 시대를 한 보 앞당길 수 있을 것으로 보인다. 무인자동차에 탑재되는 인공지능이 도로 교통에 대한 기본적인 자료와 함께 다양한 주행 데이터와 사고 데이터 그리고 지도에 관한 것들을 습득하고, 최적의 운전상태로 만들면서 그 이상을 보여줄 수 있을 것으로 기대된다.
